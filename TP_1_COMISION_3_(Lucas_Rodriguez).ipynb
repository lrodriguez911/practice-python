**EJERCICIO 3 CICLO FOR**


Solicitar al usuario que ingrese 10 números. Calcule la suma de todos los números ingresados y  cuente cuántos de ellos son par. Mostrar la suma de todos los numeros y mostrar cuanto de ellos son par.


```python
#Escribe aqui abajo la resolucion del EJERCICIO 3:
lista3 = []
for i in range(10):
  try:
    lista3.append(int(input("ingrese un numero: ")))
  except ValueError:
    print("debe ingresar un numero")
    lista3.append(int(input("ingrese un numero: ")))
print(f'la suma total de los numeros ingresados es: {sum(lista3)}')
print(f'la cantidad de numero pares es: {len([x for x in lista3 if x % 2 == 0])}')
```
